```python
from matplotlib import pylab
from random import random
```

```python
pylab.plot([0, -11.5, 11.5, 0], [90, 70, 70, 90], "g")
pylab.plot([0, -17., 17., 0], [75, 45, 45, 75], "g")
pylab.plot([0, -20, 20, 0], [55, 20, 20, 55], "g")
pylab.plot([-7, -7, 7, 7, -7], [20, 0, 0, 20, 20], "y")
pylab.text(-1, 90, '*')
pylab.text(-5, 25, "Merry Christmas!")

for i in range(0, 90):
    x = 100.*random()-50.
    y = 90*random()
    pylab.text(x, y, ':')

pylab.ylim(100, 0)
pylab.xlim(-50, 50)
pylab.show()
```
